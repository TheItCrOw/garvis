```python
from langchain_core.tools import tool
from langchain.agents import create_agent
from langchain_openai import ChatOpenAI
from langchain_ollama import ChatOllama
from langchain_core.messages import HumanMessage
from langchain_community.tools import DuckDuckGoSearchRun
from IPython.display import Image, display
from datetime import datetime
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
@tool
def weather(city:str)->str:
    """Get weather in a given city"""
    return f"The weather in {city} is sunny"
```

```python
tools = [weather, DuckDuckGoSearchRun()]
```

```python
llm_openai = ChatOpenAI(model=os.getenv("OPENAI_MODEL"), temperature=0)
```

```python
llm_ollama = ChatOllama(
    model="qwen2.5:7b",
    validate_model_on_init=True,
)
```

```python
date = datetime.now().strftime("%Y-%m-%d")
```

```python
agent_openai = create_agent(
    model = llm_openai
    , tools = tools
    , name = "weather_agent"
    , system_prompt = "You are my AI assistant that has access to certain tools. Use this tools to answer my queries to the best of your abilities"
)
```

```python
agent_ollama = create_agent(
    model = llm_ollama
    , tools = tools
    , name = "weather_agent"
    , system_prompt = "You are my AI assistant that has access to certain tools. Use this tools to answer my queries to the best of your abilities"
)
```

```python
query = f"What is the weather in Manila? Then, tell me the latest news about the stocks of Apple as of `{date}`. Also tell me a list of tools you have."
result_openai = agent_openai.invoke({"messages":[HumanMessage(content=query)]})
```

```python
for msg in result_openai["messages"]:
    if(hasattr(msg,"tool_calls")):
        for index, tool_call in enumerate(msg.tool_calls):
            print("*"*100)
            print(f"Tool Call: #{index+1}| Name: {tool_call["name"]}| Args: {tool_call["args"]}")
```

```python
print(result_openai["messages"][-1].content)
```
