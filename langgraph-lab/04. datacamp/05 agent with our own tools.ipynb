```python
from dotenv import load_dotenv
from langchain_core.tools import tool
from langchain.agents import create_agent
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
from langchain_community.tools import DuckDuckGoSearchRun
from IPython.display import Image, display
from datetime import datetime
import os
load_dotenv()
```

```python
@tool
def weather(city:str)->str:
    """Get weather in a given city"""
    return f"The weather in {city} is sunny"
```

```python
tools = [weather]
```

```python
llm = ChatOpenAI(model=os.getenv("OPENAI_MODEL"), temperature=0)
```

```python
agent = create_agent(
    model = llm
    , tools = tools
    , name = "weather_agent"
    , system_prompt = "You are my AI assistant that has access to certain tools. Use this tools to answer my queries to the best of your abilities"
)
```

```python
query = "What is the weather in Manila?"
result = agent.invoke({"messages":[HumanMessage(content=query)]})
```

```python
for msg in result["messages"]:
    if(hasattr(msg,"tool_calls")):
        for index, tool_call in enumerate(msg.tool_calls):
            print("*"*100)
            print(f"Tool Call: #{index+1}| Name: {tool_call["name"]}| Args: {tool_call["args"]}")
```

```python
print(result["messages"][-1].content)
```

```python
@tool
def social_media_follower_count(social_media:str)->str:
    """
    Retrieves the current follower count for a specific social media platform.
    
    Args:
        social_media(str): The name of the social media platform (eg. twitter, instagram, linkedin, etc)
        
    Returns:
        str: the tollower count as a formatted string
        
    Example:
        social_media_follower_count('twitter') -> '1,234 followers'
    """
    return f"You have 9876 followers on {social_media}!"
```

```python
tools = [weather,social_media_follower_count]
```

```python
llm = ChatOpenAI(model=os.getenv("OPENAI_MODEL"), temperature=0)
```

```python
agent = create_agent(
    model = llm
    , tools = tools
    , name = "general_agent"
    , system_prompt = "You are my AI assistant that has access to certain tools. Use this tools to answer my queries to the best of your abilities"
)
```

```python
query = "What is the weather in Manila and how many followers to I have in Tiktok??"
result = agent.invoke({"messages":[HumanMessage(content=query)]})
```

```python
for msg in result["messages"]:
    if(hasattr(msg,"tool_calls")):
        for index, tool_call in enumerate(msg.tool_calls):
            print("*"*100)
            print(f"Tool Call: #{index+1}| Name: {tool_call["name"]}| Args: {tool_call["args"]}")
```

```python
print(result["messages"][-1].content)
```
